```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
#read data base
data_base = pd.read_csv("StudentsPerformance.csv")

#dimension
data_base.shape
```

```python
data_base.head()
```

```python
#null fields
null = data_base.isnull()

plt.figure(figsize = (16,5))
plt.title("Analysing null fields")
sns.heatmap(null, cbar = False);
```

```python
null.sum()
```

```python
#uniques
data_base.nunique()
```

```python
#checking data duplicate
data_base.duplicated().sum()
```

```python
#statistics
data_base.describe()
```

```python
#gender analysis
data_base["gender"].value_counts(normalize = True) * 100
```

```python
f, axes = plt.subplots(1, 3, figsize = (25,5))
plt.subplots_adjust(wspace=0.5)
sns.boxplot(x = "math score", y = "gender",  data = data_base, ax = axes[0])
sns.boxplot(x = "reading score", y = "gender", data = data_base,ax = axes[1])
sns.boxplot(x = "writing score", y = "gender", data = data_base,ax = axes[2]);
```

```python
data_base.groupby(by = ["gender"])[["math score", "reading score", "writing score"]].describe().reset_index().T
```

```python
#parental level of education analysis
data_base["parental level of education"].value_counts(normalize = True) * 100
```

```python
f, axes = plt.subplots(1, 3, figsize = (20,8))
plt.subplots_adjust(wspace=0.8)
sns.boxplot(x = "math score", y = "parental level of education",  data = data_base, ax = axes[0])
sns.boxplot(x = "reading score", y = "parental level of education", data = data_base,ax = axes[1])
sns.boxplot(x = "writing score", y = "parental level of education", data = data_base,ax = axes[2]);
```

```python
data_base.groupby(by = ["parental level of education"])[["math score", "reading score", "writing score"]].describe().reset_index().T
```

```python
#lunch analysis
data_base["lunch"].value_counts(normalize = True) * 100
```

```python
f, axes = plt.subplots(1, 3, figsize = (10,3))
plt.subplots_adjust(wspace=0.8)
sns.boxplot(x = "lunch", y = "math score",  data = data_base, ax = axes[0])
sns.boxplot(x = "lunch", y = "reading score", data = data_base,ax = axes[1])
sns.boxplot(x = "lunch", y = "writing score", data = data_base,ax = axes[2]);
```

```python
data_base.groupby(by = ["lunch"])[["math score", "reading score", "writing score"]].describe().reset_index().T
```

```python
#test preparation course analysis
data_base["test preparation course"].value_counts(normalize = True) * 100
```

```python
f, axes = plt.subplots(1, 3, figsize = (16,3))
plt.subplots_adjust(wspace=0.8)
sns.boxplot(x = "test preparation course", y = "math score",  data = data_base, ax = axes[0])
sns.boxplot(x = "test preparation course", y = "reading score", data = data_base,ax = axes[1])
sns.boxplot(x = "test preparation course", y = "writing score", data = data_base,ax = axes[2]);
```
